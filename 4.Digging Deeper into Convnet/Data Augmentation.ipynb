# Data augmentation
### deep learning models are "Data Hungry"
### Small Datasets lead to overfiting most of the times.
### So lets generate more data by augmentating the training samples.

```python
import os
from keras.preprocessing import image
```

```python
img_path=os.path.join('dog.77.jpg')
```

```python
from keras.preprocessing import image
img=image.load_img(img_path,target_size=(100,100))

img=image.img_to_array(img)/255.0
img.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
#when we are going to data augmentation in keras it is expecting a batch for tht.we have one image so we have to reshape it.

img=img.reshape((1,100,100,3))

```

```python
data_generator=image.ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode="nearest"
)
```

```python
i=0
for batch in data_generator.flow(img,batch_size=1):
    
    plt.figure()
    
    imgplot=plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if i==10:
        break
        
    plt.show()    

    
    
```
