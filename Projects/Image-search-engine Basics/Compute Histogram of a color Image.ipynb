```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def Histogram_Computation(img):
    img_height=img.shape[0]
    img_width=img.shape[1]
    img_channels=img.shape[2]
    
    
    Histogram=np.zeros([256,img_channels],np.int32)
    
    
    for x in range(0,img_height):
        for y in range(0,img_width):
            for c in range(0,img_channels):
                Histogram[img[x,y,c],c]+=1
    
    return Histogram                         
        
```

```python
def Plot_Histogram(Histogram):
    plt.figure()
    plt.title("Color Image Histogram")
    plt.xlabel("Intensity Level")
    plt.ylabel("Intensity Frequency")
    
    plt.xlim([0,256])
    plt.plot(Histogram[:,0],'b')
    plt.plot(Histogram[:,1],'g')
    plt.plot(Histogram[:,2],'r')
    plt.show()
```

```python
Input_Image=cv.imread("dog.90.jpg")

print(Input_Image.shape)

Histogram=Histogram_Computation(Input_Image)

# for i in range(0,Histogram.shape[0]):
#     for j in range(0,Histogram.shape[1]):
#          print("Histogram ",i," ",j,Histogram[i][j])
           
        
Plot_Histogram(Histogram)        
            
```
